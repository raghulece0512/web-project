```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load or create a dataset (synthetic example)
# Replace this with real patient data (ensuring privacy and compliance with regulations)
data = {
    "Age": [25, 45, 35, 50, 23, 40, 33, 60],
    "BMI": [22.5, 27.3, 30.1, 25.6, 20.2, 28.7, 29.0, 32.4],
    "Smoking": [0, 1, 0, 1, 0, 1, 0, 1],  # 0 = Non-smoker, 1 = Smoker
    "Exercise": [1, 0, 1, 0, 1, 0, 1, 0],  # 1 = Regular, 0 = Irregular
    "Diabetes": [0, 1, 1, 0, 0, 1, 1, 1],  # Target variable
}

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Features and target variable
X = df.drop("Diabetes", axis=1)  # Features: Age, BMI, Smoking, Exercise
y = df["Diabetes"]  # Target: Diabetes (0 = No, 1 = Yes)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Predict for a new patient
new_patient = [[30, 26.5, 0, 1]]  # Example: Age, BMI, Smoking, Exercise
prediction = model.predict(new_patient)

print("\nDisease Prediction for New Patient:")
print("Diabetes" if prediction[0] == 1 else "No Diabetes")
```
